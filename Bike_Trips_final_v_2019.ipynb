# Introduction & Scenario
##### This case study explores the historical trip data of Cyclistic, a fictional bike-share company in Chicago. The objective is to understand how annual members (Subscribers) and casual riders (Customers) use Cyclistic bikes differently. These insights will help design a marketing strategy to convert casual riders into annual members, securing the company's future growth.

### Step 1 : Importing required libraries

```python
import pandas as pd
from datetime import datetime
```

### Step 2 : Loading our datasets

```python
q1 = pd.read_csv(r'C:\Users\platw\Downloads\Bike Scenario Case\Project\dataset\divvy_trips_2019\Divvy_Trips_2019_Q1.csv')
q2 = pd.read_csv(r'C:\Users\platw\Downloads\Bike Scenario Case\Project\dataset\divvy_trips_2019\Divvy_Trips_2019_Q2.csv')
q3 = pd.read_csv(r'C:\Users\platw\Downloads\Bike Scenario Case\Project\dataset\divvy_trips_2019\Divvy_Trips_2019_Q3.csv')   
q4 = pd.read_csv(r'C:\Users\platw\Downloads\Bike Scenario Case\Project\dataset\divvy_trips_2019\Divvy_Trips_2019_Q4.csv')
```

### Looking for compatible columns for combining them together.

```python
print(q1.columns)
print(q2.columns)
print(q3.columns)
print(q4.columns)
```

## Step 3 : Transformation. 
### As we can see, I need to make some adjustments first!

```python
q2.rename(columns=
          {'01 - Rental Details Rental ID': 'trip_id',
            '01 - Rental Details Local Start Time': 'start_time', 
            '01 - Rental Details Local End Time': 'end_time',
              '01 - Rental Details Bike ID': 'bikeid',
                '01 - Rental Details Duration In Seconds Uncapped': 'tripduration',
                  '03 - Rental Start Station ID': 'from_station_id', 
                  '03 - Rental Start Station Name': 'from_station_name',
                    '02 - Rental End Station ID': 'to_station_id', 
                    '02 - Rental End Station Name': 'to_station_name',
                      'User Type': 'usertype', 
                      'Member Gender': 'gender', 
                      '05 - Member Details Member Birthday Year': 'birthyear'}, inplace=True)
```

### Now I'm ready to concatenate them. After that, I will explore the dataset for duplicates, corrupt data, data types, etc.

```python
df = pd.concat([q1, q2, q3, q4], ignore_index=True)
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.isna().sum())
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df['start_time'] = pd.to_datetime(df['start_time']).astype('datetime64[ns]')
df['end_time'] = pd.to_datetime(df['end_time']).astype('datetime64[ns]')
```

```python
df.rename(columns={'tripduration': 'trip_duration_mins'}, inplace=True)
```

```python
df['trip_duration_mins'] = df['trip_duration_mins'].astype(str).str.replace(',','').astype(float) / 60
df['trip_duration_mins'] = df['trip_duration_mins'].round(2)
```

```python
df.head()
```

### Dropping unnecessary columns that have no impact on my analysis

```python
df.drop(columns=['bikeid','from_station_id','to_station_id','gender','birthyear'], inplace=True)
df.columns
```

```python
df.loc[df['start_time'] > df['end_time']].count()
```

#### Trips with invalid start and end time are excluded from the dataset.

```python
df.drop(df[df['start_time'] > df['end_time']].index, inplace=True)
```

### Adding columns and changing necessary data types that I will need for visualization later on.

```python
df['day_of_week'] = pd.to_datetime(df['start_time']).dt.dayofweek +1
```

```python
df['name_of_day'] = pd.to_datetime(df['start_time']).dt.day_name()
```

```python
df['month'] = pd.to_datetime(df['start_time']).dt.month_name()
df['month_number'] = pd.to_datetime(df['start_time']).dt.month
```

#### Optimize memory usage for categorical columns

```python
df['usertype'] = df['usertype'].astype('category')
```

```python
df.head()
```

```python
df['season'] = df['month'].map({'January': 'Winter', 'February': 'Winter', 'March': 'Spring', 'April': 'Spring', 'May': 'Spring', 'June': 'Summer', 'July': 'Summer', 'August': 'Summer', 'September': 'Fall', 'October': 'Fall', 'November': 'Fall', 'December': 'Winter'})
```

### Dropping any rows that are less than 1 minute and exceeding 1440 minutes (24 hours). In this case, we have 1,848 rows that are above 24 hours of use, which means that the users forgot to end their session or that the bikes have been stolen.

```python
df.loc[df['trip_duration_mins'] > 1440].count()
```

```python
df.loc[df['trip_duration_mins'] < 1].count()
```

```python
df.drop(df[df['trip_duration_mins'] > 1440].index, inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
print(df['trip_duration_mins'].min())
print(df['trip_duration_mins'].max())
```

```python
df['hour_of_day'] = pd.to_datetime(df['start_time']).dt.hour
```
